# 9.3 딥러닝의 미래

이 절은 연구 분야 종사자들의 시야를 넓혀주기 위한 목적으로 작성되었으며 딥러닝의 미래에 대한 저자의 개인적인 생각을 담고 있음

### 저자가 기대하는 딥러닝의 주요 방향
- 모델은 범용 목적의 컴퓨터 프로그램에 가까워짐
    - 현재의 미분가능 층보다 더 뛰어난 모듈로 만들어짐
    - 현재 모델의 근본적인 약점인 부족한 **추상**과 **추론**을 얻는 방법이 됨
    

- 미분 가능한 변환에서 탈피한 모델이 등장 (새로운 형태의 학습이 생겨날 것임)


- 엔지니어는 모델에 덜 관여하게 됨
    - 하이퍼파라미터 튜닝은 엔지니어의 역할이 아님
    

- 더 훌륭하고 체계적으로 이전에 학습한 특성과 구조를 사용
    - 재사용 가능하고 모듈화된 프로그램 서브루틴을 사용한 메타러닝 시스템(meta-learning system)과 유사
    

참고) 위 내용은 지도학습에 국한되지 않으며 어느 형태의 머신러닝에도 적용됨

## 9.3.1 프로그램 같은 모델

### 머신러닝 분야에서 기대하는 필수적인 변화
- 순수한 **패턴인식**을 수행하고 **지역 일반화**만 얻을 수 있는 모델을 탈피하는 것
- **추상**과 **추론**을 통해 **궁극 일반화**를 달성할 수 있는 모델이 되어야 함


### 프로그래밍 요소를 갖춘 신경망
- 프로그래밍 요소를 갖춘 신경망은 많은 프로그래밍 요소를 포함하고 있어 모델이 이를 자유롭게 조합하여 함수를 확장할 수 있음
    - 프로그래밍 요소의 예) `if` 조건문, `while`문, 변수 생성, 장기메모리를 위한 디스크 공간, 정렬 연산, 고급 자료구조(리스트, 그래프, 해시테이블) 등
    - 개발자에 의해 하드코딩 된다는 의미가 아니라 모델이 스스로 프로그래밍 요소를 조합한다는 의미
    
    
- 프로그래밍 요소를 갖춘 신경망은 현재 딥러닝 모델보다 훨씬 광범위한 공간을 표현할 수 있으며 뛰어난 일반화 성능을 가질 수 있음 (추론과 추상이 가능)


### 알고리즘 서브루틴과 기하학적 서브루틴

미래의 학습된 프로그램에는 알고리즘 모듈과 기하학적 모듈이 합쳐지게 될 것임

- 알고리즘 서브루틴(algorithmic subroutine)
    - 추론과 추상 능력을 가진 형식이 있는 알고리즘 모듈(수동으로 만든 소프트웨어)
    

- 기하학적 서브루틴(geometric subroutine)
    - 형식이 있는 직관과 패턴 인식 능력을 제공하는 기하학적 모듈(딥러닝)

### 프로그램 합성(program systhesis)

프로그램 합성은 탐색 알고리즘을 사용해 거대한 프로그래밍 가능 공간을 탐색하여 자동으로 프로그램을 합성함

- 머신러닝 방식과의 공통점
    - 입출력 쌍으로 제공되는 필요 사양을 만족시키면 탐색을 중단
    
    
- 머신러닝 방식과의 차이점
    - 하드코딩된 프로그램(신경망)에서 학습하는 것이 아님
    - 이산적인 탐색 과정(discrete search process)을 통해 소스코드를 생성하는 것

### 딥러닝과 프로그램 합성을 아우르는 분야의 등장

미래에는 프로그래밍 언어로 프로그램을 생성하는 대신에 `for`루프 등 다양한 알고리즘 요소를 갖춘 신경망을 생성하게 될 것임
- 이는 직접 소스코드를 생성하는 것 보다 훨씬 쉽고 유용함


이를 통해 머신러닝으로 풀 수 있는 문제의 범위가 크게 확장될 것임
- 적절한 훈련 데이터가 주어졌을 때 자동으로 생성할 수 있는 프로그램 공간의 확장을 의미함

## 9.3.2 역전파와 미분 가능 층을 넘어서

### 미래의 머신러닝 모델과 학습

미래의 머신러닝 모델은 프로그래밍적 요소를 가진 알고리즘 서브루틴과 기하학적 서브루틴을 모두 가지게 됨

- 기하학적 서브루틴의 경우 미분 가능한 시스템이며 경사하강법과 역전파를 사용한 학습이 가능함
- 알고리즘 서브루틴의 경우 미분 가능하지 않은 시스템이므로 이를 학습할 방법이 필요함
    - 현재는 유전 알고리즘, 진화 전략, 일부 강화 학습, ADMM(Alternating Direction Method of Multipliers)같은 방법이 있음

따라서 미래에는 그래디언트를 사용하지 않는 효율적인 탐색 과정을 거쳐 전체적으로 미분 가능하지 않은 모델을 훈련시키는 방식이 나오게 될 것임

### 모듈화와 계층화

역전파는 end-to-end로 연속된 변환의 학습에는 뛰어나지만 심층 네트워크의 모듈성을 활용하지 않으므로 계산 효율성이 떨어짐

이를 효율적으로 하기 위한 보편적인 방법은 모듈화와 계층화를 하는 것
- 동기화 메커니즘을 추가하여 훈련 모듈을 독립적으로 분리하면 역전파를 더 효율적으로 수행할 수 있음
- 이는 딥마인드의 synthetic gradient에 관한 연구에 일부 반영되었음


### 정리

- 미래에는 전체적으로 미분 가능하지 않은 모델을 훈련시키는 방식이 나오게 될 것임
- 미분 가능한 부분은 높은 효율의 역전파 방식을 사용하고 그래디언트의 이점을 살려 더 빨리 훈련될 것임

## 9.3.3 자동화된 머신러닝

### 하이퍼파라미터 튜닝의 자동화

미래에는 모델 구조가 엔지니어에 의해 고안되는 것이 아니라 학습될 것임

하이퍼파라미터 튜닝은 간단한 탐색 절차임
- 엔지니어가 원하는 바를 튜닝되는 네트워크의 손실 함수로 정의할 수 있음
- 이는 모델 파라미터 튜닝을 담당하는 AutoML 시스템에서 널리 사용되는 방식
    
    
하이퍼파라미터 튜닝 자동화 시스템은 층의 개수와 순서, 유닛과 필터의 수를 튜닝함
- AutoML과 `Hyperopt`과 같은 라이브러리가 하는 방식
- 좀 더 대담하게는 강화학습이나 유전 알고리즘 등을 사용해 아예 처음부터 적절한 구조를 학습할 수도 있음
  

참고) 데이터 전처리는 도메인 지식과 엔지니어의 목적에 대한 명확한 고수준의 이해가 필요하므로 자동화하기 어려움


### 모델 구조의 학습
또 다른 중요한 AutoML의 방향은 모델의 가중치와 함께 모델 구조를 학습하는 것
- 조금씩 다른 구조를 실험할 때 마다 새로운 모델을 훈련하는 것은 매우 비효율적임
- 강력한 AutoML 시스템은 모델의 특성이 훈련 데이터에서 역전파로 튜닝됨과 동시에 구조를 찾을 것임


### 머신러닝 엔지니어의 역할
이러한 자동화가 된다면 머신러닝 엔지니어는 가치 창조 사슬의 위로 이동하게 됨

비즈니스 목표를 잘 반영하는 손실함수를 만들고 모델이 사용될 환경(또는 사용하는 사용자)에 미칠 영향을 이해하는데 더 많은 노력을 기울이게 될 것임

## 9.3.4 영구 학습과 모듈화된 서브루틴 재사용

### 재사용성

미래에는 모델의 재사용성이 증가하게 될 것임

- 많은 데이터셋은 새로운 복잡한 모델을 처음부터 만들기에 충분하지 않으며 이전의 데이터셋의 정보를 재사용할 필요가 있음
- 새로운 작업과 기존 작업 사이에 중복되는 점이 많다면 새로운 모델의 훈련은 비효율적임

현재는 동일한 공통 기능을 수행하는 부분 모델의 가중치를 재사용 하는 방식이 있음 (5장의 vgg16모델의 재사용과 같은 방식)

미래에는 재사용성이 더 일반화 되어 모델의 구조와 훈련 과정까지도 재사용하게 될 것임

또한 모델이 프로그램과 더욱 비슷해지면 함수나 클래스처럼 모델의 **프로그램 서브루틴**을 재사용하기 시작할 것임

### 추상화

미래에는 재사용 가능한 요소(서브루틴들)을 활용하여 특정 작업을 위한 모델을 빠르게 개발할 수 있는 메타 학습기가 등장하게 될 것임

- 메타 학습 시스템이 전체 공통 라이브러리에서 고수준의 재사용 가능한 블록(서브루틴)을 바꾸어 가며 새로운 프로그램을 조합
- 프로그램 서브루틴을 개발할 때는 추상화된 재사용 가능한 서브루틴을 만들고 전체 공통 라이브러리에 저장
- 이러한 서브루틴은 기하학적(사전 훈련된 표현을 가진 딥러닝 모델)이거나 알고리즘적(소프트웨어 엔지니어가 다루는 것과 비슷한 라이브러리)인 요소일 수도 있음

이러한 과정은 궁극 일반화를 달성하기 위해 필요한 **추상화**를 구현하게 됨

## 9.3.5 장기 비전

### 머신 러닝에 대한 장기 비전 요약

- 모델은 알고리즘 모듈과 기하학적 모듈을 혼합한 형태로 더 프로그램 같아질 것임
    - 형식 추론, 탐색, 추상화 능력을 제공하는 알고리즘 모듈
    - 비형식적인 직관과 패턴 인식 능력을 제공하는 기하학적 모듈
    - 이러한 모델은 더 강한 일반화 성능을 가지게 됨
    

- 모델은 엔지니어가 하드코딩하는 대신 자동으로 만들어질 것임


- 서브루틴의 재사용을 통해 추상화를 구현하게 됨
    - 메타 학습 시스템에 의해 전체 공통 라이브러리에서 서브루틴을 재사용 하게 됨
    - 자주 발견되는 문제 해결 패턴은 서브루틴으로 바뀌어 전체 공통 라이브러리에 저장됨


- 전체 공통 라이브러리와 연계된 모델 성장 시스템은 사람과 비슷한 궁극 일반화의 어떤 형태를 달성하게 될 것임
