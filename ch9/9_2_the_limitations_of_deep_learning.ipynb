# 9.2 딥러닝의 한계

### 딥러닝의 한계

현재의 딥러닝 기술로 해결할 수 없는 경우는 다음과 같음

- 프로그래밍 처럼 논증이 필요한 경우
- 장기 계획을 세워 과학적 방법을 적용하는 경우
- 알고리즘을 사용하여 데이터를 조작하는 일


그 이유는 딥러닝 모델은 한 벡터공간을 다른 벡터공간으로 매핑하기 위한 단순하고 연속된 기하학적 변환들의 연결일 뿐이기 때문
- 한 매니폴드 $X$에서 $Y$로 매핑하는 것이 딥러닝이 할 수 있는 것의 전부임

**현재의 딥러닝은 연속된 기하학적 변환으로 나타낼 수 없는 근본적인 문제를 해결할 수 없음**

## 9.2.1 머신러닝 모델의 의인화 위험

현대 AI에서 오는 위험은 딥러닝 모델이 하는 일을 사람이 하는 것처럼 과대평가 하는 것임

특히 모델이 잘못 분류하도록 고안된 적대적인 샘플(Adversarial example)의 예를 보면 이런 문제를 잘 이해할 수 있음
- 아래 그림처럼 판다 이미지에 긴팔원숭이 그래디언트를 더하면 판다를 긴팔원숭이로 분류하게 됨 (경사 상승법을 사용해 클래스의 예측이 최대가 되도록 이미지를 수정함)


<img src="./images/adversarial_example.png" alt=adversarial_example align="left" />

위 예를 통해 딥러닝 모델이 불안정하다는 것을 알 수 있음
- 딥러닝 모델의 입력-출력 매핑과 사람의 지각은 큰 차이가 있음
- 딥러닝 모델은 입력을 전혀 이해하지 못함(사람과 같은 방식으로 입력을 이해할 수 없다는 의미)

머신러닝과 사람의 학습의 근본적인 차이는 다음과 같음
- 사람은 지각을 통해 학습하고 추상적인 개념을 형성
- 머신러닝은 많은 양의 훈련 샘플을 통해 사람에게 형성된 추상적인 개념에 매핑하는 기하학적 변환을 학습하고 모델을 형성함
    - 이 매핑은 사람의 경험으로 만들어진 추상적인 개념(사람 마음속에 내재된 진짜 모델)을 단순히 흉내낸 것
    

따라서 머신러닝 기술자는 항상 아래의 내용을 기억하고 있어야 함

- 신경망이 수행하는 작업을 '이해'한다고 믿는 함정에 빠지면 안됨
- 신경망은 적어도 사람이 느끼는 방식으로는 이해하지 못함

## 9.2.2 지역 일반화 vs. 궁극 일반화

참고) 지역 일반화와 궁극 일반화라는 단어는 이 책의 저자가 딥러닝과 사람을 구분하기 위해 사용한 단어


### 궁극 일반화

사람은 경험하지 못한 어떤 것을 표현하기 위해 알고 있는 개념을 합칠 수 있음
   - 예) 청바지를 입은 말이 있는 사진을 떠올리는 것, 복권에 당첨되면 무엇을 할지 상상하는 것 등    

가상의 일을 다루는 이런 능력은 **추상(abstraction)**과 **추론(reasoning)**을 통해 마음속에 있는 모델의 공간을 직접 경험할 수 있는 것 이상으로 확장시켜줌

이는 사람이 가진 인지 능력의 특징이며 이를 **궁극 일반화(extreme generalization)**라고 함
- 데이터가 조금만 있거나 심지어 새로운 데이터가 전혀 없어도 이전에 경험한 적 없는 새로운 상황에 적응하는 능력


### 지역 일반화

심층 네트워크가 수행하는 입력-출력 매핑은 새로운 데이터가 훈련할 때 보았던 것과 조금만 달라도 납득할 만한 결과를 내지 못함

이는 심층 네트워크의 일반화의 특징이며 이를 **지역 일반화(local generalization)**라고 함

### 요약

- 현재의 딥러닝 모델은 과거 데이터와 비슷한 환경에만 적용되는 지역 일반화만 수행할 수 있음
- 사람의 인지 능력은 완전히 새로운 환경에 빠르게 적응하고 먼 미래의 상황을 계획할 수 있는 궁극 일반화를 수행할 수 있음

## 9.2.3 정리

- 지금까지의 딥러닝의 실제 성공은 연속된 기하학적 변환을 사용한 공간 $X$에서 공간 $Y$로 매핑하는 능력
    - 이는 모든 산업을 획기적으로 변화시킬 수 있음
    - 하지만 사람 수준의 AI에는 한참 멀었음
    

- 딥러닝이 9.2절에서 다룬 한계를 극복하고 사람의 뇌와 경쟁하려면 단순 입출력 매핑을 벗어나 **추상**과 **추론**을 해야함
